```python
import pandas as pd
```

```python
# Parks Location (CSV)
spark.read.option("header", True).option("sep", ";") \
    .csv("Files/raw/static/parksMap/parks-polygon-representation.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_parks_location")

# Park amenities (Excel)
amenity_df = pd.read_excel("/lakehouse/default/Files/raw/static/parksAmenities/parkAmenityTable.xlsx", header=0)
amenity_spark_df = spark.createDataFrame(amenity_df)
amenity_spark_df.write.format("delta").mode("overwrite").saveAsTable("raw_park_amenities")

# Events (Excel)
events_df = pd.read_excel("/lakehouse/default/Files/raw/events/parkEvents.xlsx", header=0)
events_spark_df = spark.createDataFrame(events_df)
events_spark_df.write.format("delta").mode("overwrite").saveAsTable("raw_events")

# Holidays (CSV)
spark.read.option("header", True).csv("Files/raw/holidays/bc_holidays.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_holidays")

# Weather Historical
spark.read.option("header", True).csv("Files/raw/time_series/weatherVancouver.csv/part-*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_weather_historical")

# Weather Forecast 
spark.read.option("header", True).csv("Files/raw/time_series/weatherForecast.csv/*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_weather_forecast")

# Popular times (CSV, with part files)
spark.read.option("header", True).csv("Files/raw/time_series/parksUsualTimes.csv/part-*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_popular_times")

# Live times (CSV, with part files)
spark.read.option("header", True).csv("Files/raw/time_series/parksLiveTimes.csv/part-*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_live_times")

# Parks ID
spark.read.option("header", True).csv("Files/raw/static/parksID/part-*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_parks_id")

# Parks Observational Study
spark.read.option("header", True).csv("Files/raw/time_series/parksObservationalStudy.csv/part-*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_parks_observational_study")
spark.read.option("header", True).csv("Files/raw/time_series/parksObservationalStudyID.csv/part-*.csv") \
    .write.format("delta").mode("overwrite").saveAsTable("raw_parks_observational_study_id")

# Parks classifications
df = spark.read.option("header", True).csv("Files/raw/static/parksClassifications/part-*.csv")

for col_name in df.columns:
    safe_col_name = col_name.strip().replace(' ', '_').replace('/', '_').replace('(', '').replace(')', '')
    df = df.withColumnRenamed(col_name, safe_col_name)

df.write.format("delta").mode("overwrite").saveAsTable("raw_parks_classifications")
```
